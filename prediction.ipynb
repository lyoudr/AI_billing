```python
import matplotlib.pyplot as plt 
from scipy import stats

"""
service_name : number mapping
mappings = {
  'App Engine': 1,
  'Compute Engine': 2,
  'Networking': 3,
  'BigQuery': 4
}
"""
# ------------------------------------- Linear Regression --------------------------------
x = [1, 2, 3, 4]
y = [0.0, 0.04, 0.26, 0.0]
slope, intercept, r, p, std_err = stats.linregress(x, y)
print('first relationship r is ->', r)
# Firstly, I guess it is linear regression model
# Mymodel
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
# Draw the original scatter plot:
# plt.scatter(x, y)
# Draw the line of linear regression:
# plt.plot(x, mymodel)

# Predict Future Values
# I input a service , such as "App Engine"
future_cost = myfunc(2)
print('first future_cost is ->', future_cost) 

# Since the r : the relationship between the values of the x-axis and the values of the y-axis 
# is very low, only 0.22. I chose another model to predict data


# ----------------------------------- Polynomial Regression ---------------------------------
import numpy 
from sklearn.metrics import r2_score

x = [1, 2, 3, 4] # represent service name
y = [0.0, 0.04, 0.26, 0.0] # represent cost in USD

# NumPy has a method that lets us make a polynomial model:
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))
# Then specify how the line will display, we start at position 1, and end at position 4
myline = numpy.linspace(1, 4, 4)
# Draw the original scatter plot:
plt.scatter(x, y)
# Draw the line of polynomial regression:
plt.plot(myline, mymodel(myline))

# The same still want to find the relationship between x and y 
# using R-square
print('second relationship r2_score is ->', r2_score(y, mymodel(x)))

# from the r-squared value is equal to 1 , we see this model is very suitable for predicting future values
# For example, we can predict the future cost of "Compute Engine"
future_cost = mymodel(2)
print('second future_cost is ->', future_cost)
```
